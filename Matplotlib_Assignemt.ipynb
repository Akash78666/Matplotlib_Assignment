```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.
```

```python
Ans:-
you can use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
colours=np.random.rand(50)
size=1000* np.random.rand(50)

plt.scatter(x,y  ,c=colours,s=size, alpha=.8)
plt.xlabel("this my x-data")
plt.ylabel("this my y-data")
plt.title("sri Ram")
plt.show()
```

```python
Ans:-subplots() function can be used to create a figure with multiple subplots. It's default is to create a single (sub)plot, but we can create more subplots by passing arguments indicating the number of rows and columns of subplots we want (thinking of subplots as a 2D grid)
```

```python

import matplotlib.pyplot as plt
import numpy as np
#plot1
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])


plt.subplot(2,1,1)
plt.plot(x,y)

#plot2
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,1,2)
plt.plot(x,y)


```

```python
import matplotlib.pyplot as plt
import numpy as np

x =  np.array(["Apple", "Microsoft", "Google", "AMD"])
y =  np.array([3000, 8000, 1000, 10000])
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("pw")
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(100, 10, 200)
y = np.random.normal(90, 20, 200)
plt.xlabel("box1")
plt.ylabel("box2")
plt.title("pw2")
plt.box(x,y)
plt.show()
```
